```python
names_all = df["Name"]

# convert the series into an array
names_all = names_all.values
names_all = set(names_all)
len(names_all)
# print duplciate names in names_all
# import collections
# print([item for item, count in collections.Counter(names_all).items() if count > 1])


```

Documented APIs

```python
f = open("docAPIs.txt", "r")
names_doc = f.read().splitlines()
f.close()
# len(names_doc)
# names_doc = set(names_doc)
len(names_doc)
```

```python


# print all names that are present in names_all but not present in names_doc
undocumented = []
for name in names_all:
    if name not in names_doc:
        undocumented.append(name)

len(undocumented)
```

```python

# write the undocumented APIs to a file
f = open("undocumentedAPIsStrict.txt", "w")
for name in undocumented:
    f.write(name + "\n")
f.close()
```
