```python
import pandas as pd 
import json
import re
import ast
```

```python
df = pd.read_csv('WechatApisParaupdated.csv')
df
```

```python
def template_based_value(parameter_str):
    if parameter_str == "{}":
        return "{}"
    try:
        parameters = json.loads(parameter_str.replace("'", '"'))
    except:
        parameters = {}
    
    template = {}
    for key, value in parameters.items():
        value_type = value.lower()
        if value_type == 'int':
            template[key] = 1
        elif value_type == 'float' or value_type == 'double':
            template[key] = 1.0
        elif value_type == 'long':
            template[key] = 1
        elif value_type == 'boolean':
            template[key] = True
        elif value_type == 'string':
            template[key] = 'test'
        elif value_type == 'jsonobject':
            template[key] = {}
        elif value_type == 'jsonarray':
            template[key] = []
        else:
            print(f"Unknown type: {value_type} for {key}")
            template[key] = 'test'  # Default case for unknown types
    return json.dumps(template)
```

```python
df['Template'] = df['Parameters'].apply(template_based_value)
# df
```

```python
selected_columns = df[['Name', 'Template']]
selected_columns.tail()


with open('./output1.txt', 'w') as file:
    for index, row in selected_columns.iterrows():
        if pd.notna(row['Template']):
            try:
                template_dict = json.loads(row['Template'].replace("'", "\""))
            except json.JSONDecodeError:
                template_dict = {}
            
            if template_dict:
                template_parts = [f"{key}:{json.dumps(value)}" for key, value in template_dict.items()]
                line = f"{row['Name']} " + ",".join(template_parts) + ","
            else:
                line = row['Name']
        else:
            line = row['Name']
        file.write(line + '\n')
```
