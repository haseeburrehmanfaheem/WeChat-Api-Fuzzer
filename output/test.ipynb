```python
import json
import pandas as pd
```

```python
# Load the JSON file
file_path = '/Users/haseeb/Desktop/devTool/output/all.json'
with open(file_path, 'r') as file:
    data = json.load(file)

# Split data into different categories
access_denied_data = {}
catch_data = {}
other_data = {}

for key, value in data.items():
    value_str = str(value)
    if 'access denied' in value_str:
        access_denied_data[key] = value
    elif 'catch' in value_str:
        catch_data[key] = value
    else:
        other_data[key] = value

# Convert to DataFrames
df_access_denied = pd.DataFrame(list(access_denied_data.items()), columns=['Function', 'Details'])
df_catch = pd.DataFrame(list(catch_data.items()), columns=['Function', 'Details'])
df_other = pd.DataFrame(list(other_data.items()), columns=['Function', 'Details'])



# print the lenght of all datafrane
print("Access Denied : ", len(df_access_denied))
print("Try-Catch : ", len(df_catch))
print("Other (worked): ", len(df_other))

# add all lengths and print
total = len(df_access_denied) + len(df_catch) + len(df_other)
print("Total: ", total)
# save df_other in a csv
df_other.to_csv('df_other.csv', index=False)

df_access_denied



```
